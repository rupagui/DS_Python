### Usando el método Open

```python
# read_csv carga todo el dataset en memoria.
# Open va línea por línea y libera memoria después de cada iteración

filename = 'customer-churn-model/Customer Churn Model.txt'
fullpath = os.path.join(mainpath,filename)
data2 = open(fullpath,'r') # r: lectura     w: escritura    rw:lectura-escritura
```

```python
# ¿Cuántas columnas tenemos?
cols=data2.readline().strip().split(',')
```

```python
cols
```

```python
ncols = len(cols)
```

```python
ncols
```

```python
# El número de filas ya no es tan fácil
# Lo 1º es crear un diccionario que tenga por claves los nombres de las columnas. Recorro las columnas con un for:

main_dict = {}
for col in cols:
    main_dict[col]=[]
```

```python
main_dict
# Obtenemos algo parecido a un JSON
```

```python
# Y ahora hay que rellenar cada columna (clave) con sus valores

counter = 0

for line in data2:    
    values = line.strip().split(',')    
    for i in range(len(cols)):   
        main_dict[cols[i]].append(values[i])   
    counter = counter + 1

print("El dataset contiene %d filas y %d columnas"%(counter,ncols))
    
```

```python
# Para cargarlo ya manualmente en un dataframe
df=pd.DataFrame(main_dict)
```

```python
df.head()
```

```python
# Vamos a hacer también un ejemplo de escritura, en el que el separador sea además el tabulador

infile = fullpath
outfile = mainpath + '/' + 'customer-churn-model/Tab Customer Churn Model.txt'
```

```python
with open(infile) as infile1:
    with open(outfile,'w') as outfile1:  # Ojo que por defecto el open es de lectura. Hay que especificar la escritura
        for line in infile1:
            fields = line.strip().split(',') 
            outfile1.write('\t'.join(fields))
            outfile1.write('\n')
        
```

```python
df = pd.read_csv(outfile, sep = '\t')
```

```python
df.head()
```

## Leer datos desde una url:

```python
medals_url = 'http://winterolympicsmedals.com/medals.csv'
```

```python
medals_data = pd.read_csv(medals_url)
```

```python
medals_data.head()
```

```python
# Importamos dos librerías que serán de utilidad

import csv # Bastantes métodos para trabajar con csv's
import urllib3 # Para navegar y acceder a información desde una url

http = urllib3.PoolManager()
r = http.request('GET', medals_url)
r.status
r.data

# Scrapeamos directamente la web:
```

```python
# Hemos obtenido un string binario, que hay que convertir a string:

response = r.data
decoded_response = response.decode()
```

```python
# troceamos primero por filas:
filas = decoded_response.strip().split('\n') 

# Extraemos cabeceras:
headers = filas[0].split(',')

# Creamos diccionario y claves:
diccionario = {}
for i in headers:
    diccionario[i]=[]

# Rellenamos de valores:
for i in range (1,len(filas)):   
    valores = filas[i].strip().split(',')    
    for i in range(len(headers)):   
        diccionario[headers[i]].append(valores[i])   

# ... Y creamos el dataframe:
df = pd.DataFrame(diccionario)
df.head()
```
